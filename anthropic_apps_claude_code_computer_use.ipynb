## Phần A — Claude Code 💻

**Claude Code** là một **AI coding agent** chạy trong terminal. Nó không chỉ trả lời câu hỏi mà **chủ động làm việc**: đọc/sửa file, chạy lệnh, chạy test, tạo commit...

### Bản chất
Claude Code = một **vòng lặp agentic** (giống bài Tool Use) với bộ công cụ mạnh:
- `bash` — chạy lệnh terminal
- `read` / `write` / `edit` — thao tác file
- `glob` / `grep` — tìm file & nội dung
- và nhiều hơn nữa

Claude tự quyết định gọi công cụ nào, theo trình tự nào để hoàn thành nhiệm vụ bạn giao.


````python
# Claude Code là CLI — cài và chạy trong TERMINAL (không phải trong notebook):
#
#   npm install -g @anthropic-ai/claude-code
#   cd duong/dan/du-an
#   claude
#
# Sau đó gõ yêu cầu bằng ngôn ngữ tự nhiên, ví dụ:
#   "Thêm validation cho form đăng ký và viết test"
#   "Tìm và sửa bug khiến trang chủ load chậm"
#
# Claude Code sẽ tự đọc code, sửa, chạy test... và xin xác nhận khi cần.
print("Claude Code chạy trong terminal — xem hướng dẫn ở ô markdown trên.")
````

### Vì sao Claude Code minh họa tốt bài học?
- **Tool Use ở quy mô lớn**: hàng chục công cụ phối hợp.
- **Agentic loop**: tự lặp đến khi xong nhiệm vụ.
- **Human-in-the-loop**: xin phép trước hành động nguy hiểm (xóa file, push...).
- Bạn có thể tự xây thứ tương tự bằng API + tool use đã học.


## Phần B — Computer Use 🖱️

**Computer Use** cho phép Claude **điều khiển máy tính như con người**: xem màn hình (ảnh chụp), di chuột, gõ phím, click. Dùng cho việc tự động hóa GUI mà không có API.

### Cách hoạt động (vòng lặp)
```
1. Bạn chụp màn hình -> gửi cho Claude
2. Claude nhìn ảnh, quyết định hành động (click ở (x,y), gõ chữ...)
3. Code của bạn THỰC THI hành động trên máy/sandbox
4. Chụp màn hình MỚI -> gửi lại
5. Lặp đến khi hoàn thành nhiệm vụ
```
